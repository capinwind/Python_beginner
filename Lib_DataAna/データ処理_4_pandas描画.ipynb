## 追記　plot

* メソッドのように使う：　　　plot(kind = 'area')
* オブジェクトのように使う：　plot.area()

plotの種類：

* 領域　　　　　　plot.area()
* 棒　　　　　　　plot.bar()
* 箱ひげ　　　　  plot.box()
* ヒストグラム　  plot.hist()
* 密度　　　　　　plot.kde()
* 散布図　　　　　plot.scatter('x', 'y')
* 円　　　　　　　plot.pie(subplots = True)
