#### ピボットテーブル

melt を使うと複数列で持つ値を行方向に展開して作成できる

```python
e = pd.melt(d, id_vars = ['species'], var_name = 'target', value_name = 'val')
e.head()
e
```

species と　target　の組ごとに、 groupbyを用いて平均値を求める。
複数キーを用いたので、indexも複数キーになっている。次の処理で必要なため、 reset_indexでキーをリセットする

```python
f = e.groupby(['species', 'target']).mean().reset_index()
f
```

#### piot

```python
f.pivot(index = 'target', columns = 'species', values = 'val')
```
