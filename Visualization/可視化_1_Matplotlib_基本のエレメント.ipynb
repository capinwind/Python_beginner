### 例1 サインカーブ

```python
x = np.linspace(0,np.pi*2)
y = np.sin(x)
plt.plot(x,y)
```

## 色・マーク・線種の変更

* Subplotの作り方。詳細は別途あり
* 色・マーク・線種をまとめて指定できる
　 * 'ro': 赤い（r）丸（o）を表す 
   * 'gD-': 緑(g) ダイヤ(D) 実線(-)
* "c=???" 色の単独指定。指定の仕方：(1)色名(navy)　(2)カラーコード(#9400D3)　(3)RGB((1.0,0.0,0.0))
* "marker=???" マークの単独指定
* "linestyle=???" 線種の単独指定

#### List of Color

* 'b' blue
* 'g' green
* 'r' red
* 'c' cyan
* 'm' magenta
* 'y' yellow
* 'k' black - kuro^^
* 'w' white

#### List of Marker

* '.'    点
* ','    ピクセル
* 'o'    丸
* '*'    ※
* 'v'    下向き△
* '^'    上向き△
* '<'    左向き△
* '>'    右向き△
* 's'    正方形
* 'p'    五角形
* 'h'    六角形１
* 'H'    六角形２
* '+'    ＋
* 'x'    X
* 'D'    菱形１
* 'd'    菱形２
* '|'    縦棒
* '_'    横棒

#### List of linestyle

* '-' 実線
* '--' 破線
* '-.' 一点鎖線
* ':' 点線

#### 例２ 色・マーク・線種の変更

```python
plt.subplot(131)
plt.plot(x,y,'ro')
plt.subplot(132)
plt.plot(x,y,'gD-')
plt.subplot(133)
plt.plot(x,np.sin(x),c='navy',marker='D')
plt.plot(x,np.sin(-x),c='#9400D3',marker='d',linestyle='-.')
plt.plot(x,np.cos(x),marker='H',c=(1.0,0.0,0.0))
```

## タイトル・軸ラベル・凡例

#### まずは例をみてみよう

* タイトルの指定→１
* ラベル、凡例の文字指定→2/3/4
* 横軸・縦軸のラベル指定→5/6
* 横軸・縦軸の表示範囲指定→７
* 各データのラベル指定→８
* 凡例の表示→９
　　* lengendによって表示させる
    * locで位置を指定できる。指定しない場合、自動計算される。
    * shadowで影を指定できる

```python
plt.title('Sin curve サインカーブ')  #1
plt.plot(x,np.sin(x),c='navy',marker='*',label='sin')  #2
plt.plot(x,np.sin(-x),c='#9400D3',linestyle='-.',label='-sin')  #3
plt.plot(x,np.cos(x),c=(1.0,0.0,0.0),label='cos')  #4
plt.xlabel('x value')  #5
plt.ylabel('y value', size = 15)  #6
plt.xlim(0, 2 * np.pi)  #7
plt.xticks(np.linspace(0, 2*np.pi,5),  #8
          ['0', 'pi/2','pi','3 pi/2', '2 pi'])
plt.legend(loc="upper right", shadow = False)  #9
```

## 日本語フォント
Matplotlibの既定値は英語のため、日本語は文字化けする

* 各関数ごとにFont指定をする(下記の例)
* プログラムの最初に宣言する(プログラムの最初)
* MatplotlibのConfigファイルを修正　(https://qiita.com/yniji/items/3fac25c2ffa316990d0c)


```python
#WINDOWS:  fp = plt.matplotlib.font_manager.FontProperties(fname=r'C:\Windows\Fonts\msgothic.ttc')  
#MAC:   fp = plt.matplotlib.font_manager.FontProperties(fname=r'/Library/Fonts/Osaka.ttf')
fp = plt.matplotlib.font_manager.FontProperties(fname=r'/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc')  # Ubuntu
plt.plot(x,np.sin(x), label='サインカーブ')
plt.xlabel('X', fontproperties=fp)
plt.legend(prop=fp)
plt.title('3', fontproperties=fp)
```
