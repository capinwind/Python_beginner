## ファイルへの保存

* 拡張子：　eps, jpeg, jpg, pdf, pgf, png, ps, raw, rgba, svg, svgz, tif, tiff  #save with bmp will show what is supported.
* 解像度：　dpi, default = 100; dot per inch

```python
from IPython.core.pylabtools import figsize
figsize(16,4)
#plt.rcParams['figure.figsize'] = 6,3

print('画像in描画領域')
plt.subplot(131)
x = np.random.rand(100)
y = np.random.rand(100)
plt.plot(x,y,'o')

plt.subplot(132)
x = np.random.rand(100)
y = np.random.rand(100)
plt.plot(x,y,'o')
plt.subplot(133)
x = np.random.rand(100)
y = np.random.rand(100)
plt.plot(x,y,'o')
plt.savefig('Matplotlibファイル保存練習.tiff', dpi = 50) #2.6Mpa@100dpi; 64Mpa@500dpi
plt.savefig('Matplotlibファイル保存練習_tight.tiff', dpi = 50, bbox_inches = 'tight') # 余白を少なく
```

## 三次元プロット

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()

figsize(6,4)

ax = fig.add_subplot(111, projection = '3d')

x = np.random.rand(100)
y = np.random.randn(100)
z = np.random.randint(100)
ax.scatter3D(x,y,z)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()

figsize(10,4)

ax = fig.add_subplot(121, projection = '3d')

x = np.random.rand(100)
y = np.random.randn(100)
z = np.random.randint(100)
ax.scatter3D(x,y,z)

ax = fig.add_subplot(122, projection = '3d')

x = y = np.arange(-3.3, 3.3, 0.3)
x,y = np.meshgrid(x, y)
z = np.cos(np.sqrt(x**2 + y**2))
ax.plot_surface(x,y,z,
               rstride = 1,
               cstride = 1,
               )
```
